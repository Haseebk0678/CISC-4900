```python
# https://medium.com/@jinilcs/a-simple-keras-model-on-my-laptop-webcam-dda77521e6a0
import cv2
import numpy as np
from PIL import Image
from keras import models
import pandas as pd

#Load the saved model
video = cv2.VideoCapture(0)
video.set(3,100)
video.set(4,100)
video.set(10,100)
input_shape = (100,100)

while True:
    _, frame = video.read()

    # frame = cv2.imread(filename)
    #Convert the captured frame into RGB
    im = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye_tree_eyeglasses.xml')
    faces = face_cascade.detectMultiScale(frame, 1.1, 4)

    # Draw rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

        
    # Display the output    
    resized = cv2.resize(im, input_shape)
    resized = np.reshape(resized, input_shape[0] * input_shape[1])
    im_df = pd.DataFrame(resized)
    #Resizing into 100x100 because we trained the model with this image size.
    X = im_df / 255.0
    X = X.values.reshape(-1,100,100,1)
    #Calling the predict method on model to predict 'me' on the image
    print(model.predict(X)[0][0])
    prediction = model.predict(X)[0][0]
    print(prediction)
#    Closer prediction is to 1, stronger guess of NO MASK, closer is to 1, stronger guess of mask
    if prediction < .5:
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    cv2.imshow("Capturing", frame)
    key=cv2.waitKey(1)
    if key == ord('q'):
            break
# video.release()
# cv2.destroyAllWindows()
```
