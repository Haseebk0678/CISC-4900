```python
import numpy as np
from tensorflow.keras.models import load_model
import cv2
import numpy as np
from PIL import Image
from keras import models
import pandas as pd
import os
```

```python
model = load_model('best_model_3.h5')
# Change to input_shape = (100,100,3) if using _2 2,3,5
input_shape = (100,100)
```

```python
# https://medium.com/@jinilcs/a-simple-keras-model-on-my-laptop-webcam-dda77521e6a0
import cv2
import numpy as np
from PIL import Image
from keras import models
import pandas as pd

#Load the saved model
video = cv2.VideoCapture(0)
input_shape = (100,100)
video.set(10,100)

while True:
    _, frame = video.read()

    # frame = cv2.imread(filename)
    #Convert the captured frame into RGB
    im = frame
    
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_alt.xml')
#     haarcascade_frontalface_alt
#     haarcascade_frontalface_default
    faces = face_cascade.detectMultiScale(frame, 1.1, 4)
    if(len(faces) > 0):
    #     print(faces)
        # Draw rectangle around the faces
        face_im = im
        for (x, y, w, h) in faces:
            face_im = im[y:y+h + 20, x:x+w + 20]
        
        resized = cv2.resize(face_im, input_shape)
        # Convert the image to a numpy array
        image_array = np.array(resized)

        # Resize the image to the correct dimensions (if necessary)
        image_array = np.resize(image_array, (1, 100, 100,3))

        # Normalize the image data
        image_array = image_array / 255.0
        # Make a prediction with the model
        prediction = model.predict(image_array)
        print(prediction)
    #    Closer prediction is to 1, stronger guess of NO MASK, closer is to 1, stronger guess of mask
        pred = prediction[0][0] * 100

        if prediction[0][0] > .6:
            label = "No Mask"
            pred = prediction[0][0]*100
            for (x, y, w, h) in faces:
                cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 255), 2)
                cv2.putText(frame, label + " " +str(pred), (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
        else:
            label = "Mask"
            pred = prediction[0][1]*100
            for (x, y, w, h) in faces:
                cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
                cv2.putText(frame, label +" "+ str(pred), (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    else:
        eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
        eyes = eye_cascade.detectMultiScale(frame, 1.1, 4,  minSize=(30, 30))
        face_im = im
        
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(frame, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)
        

        resized = cv2.resize(face_im, input_shape)
        # Convert the image to a numpy array
        image_array = np.array(resized)


        # Resize the image to the correct dimensions (if necessary)
        image_array = np.resize(image_array, (1, 100, 100,3))

        # Normalize the image data
        image_array = image_array / 255.0
        # Make a prediction with the model
        prediction = model.predict(image_array)
        print(prediction)
    #    Closer prediction is to 1, stronger guess of NO MASK, closer is to 1, stronger guess of mask
        pred = prediction[0][0] * 100

        if prediction[0][0] > .6:
            label = "No Mask"
            pred = prediction[0][0]*100
            for (ex, ey, ew, eh) in eyes:
                cv2.rectangle(frame, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)
                cv2.putText(frame, label + " " +str(pred), (ex, ey-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
        else:
            label = "Mask"
            pred = prediction[0][1]*100
            for (ex, ey, ew, eh) in eyes:
                cv2.rectangle(frame, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)
                cv2.putText(frame, label + " " +str(pred), (ex, ey-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)
    cv2.imshow("Capturing", frame)
    key=cv2.waitKey(1)
    if key == ord('q'):
            break
# video.release()
# cv2.destroyAllWindows()
```
